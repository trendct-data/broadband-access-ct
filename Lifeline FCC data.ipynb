# Lifeline enrollment data 

The FCC's Lifeline program helps poor people pay for telecommunications services -- telephone and cell phones, as well as Internet service starting in 2016. There's some uncertainty about whether the program will face changes with a new FCC chairman appointed by President Donald Trump.

Data from this independent, non-profit which oversees the program:
* http://www.usac.org/li/about/process-overview/program-stats.aspx
* https://en.wikipedia.org/wiki/Universal_Service_Fund#Universal_Service_Administrative_Company


# Loading, reviewing the data

```python
lifeline = pd.read_excel("data/lifeline/Eligible-Lifeline-Population-Statistics.xlsx",
                      header=[0,1])
```

# Participation rates by state

How many eligible people are using the program?

```python
def participation_by_state():
    df = lifeline.copy()["Lifeline Participation Rate3","Unnamed: 4_level_1"]\
    .sort_values(ascending=False).to_frame().reset_index()
    
    df.columns="st","participation_rate"
    
    return df

participation_by_state()
```

```python
def ct_only():
    df = lifeline.copy()
    
    return df.loc["CT"].to_frame()

ct_only()
```
