```python
import pandas as pd
```

```python
def get_data(year="15"):
    df = pd.read_csv("data/census/income/ACS_"+str(year)+"_1YR_B28004/ACS_"+str(year)+"_1YR_B28004_with_ann.csv",
                    skiprows=1)
    return df

data_15 = get_data(year="15")
data_15
```

```python
for c in data_15.columns:
    print c
```

```python
# data types
datum = {
    "MOE":"Margin of Error",
    "est":"Estimate"
}
    
# income buckets
inc_buckets = {
    "20K":"$10,000 to $19,999",
    "35K":"$20,000 to $34,999",
    "50K":"$35,000 to $49,999",
    "75K":"$50,000 to $74,999",
    "75+":"$75,000 or more",
    "NA":""
}

#subscription types
sub_types = {
    "dialup":" - With dial-up Internet subscription alone",
    "broadband":" - With a broadband Internet subscription",
    "none":" - Without an Internet subscription",
    "NA":""
}
```

```python
# Generate a column name
def col(dat, inc, sub):
   
    
    return datum[dat] + "; " + inc_buckets[inc] +":" + sub_types[sub]
```

```python
col("est","20K","broadband")
```

```python
def three_buckets(df):
    ret = df.copy()
    ret["COMBINED $20,000 to $49,999 with broadband"] = ret[col("est","35K","broadband")]\
    + ret[col("est","50K","broadband")]
    
    ret["COMBINED $20,000 to $49,999 pop"] = ret[col("est","35K","NA")]\
    + ret[col("est","50K","NA")]

    return ret

three_buckets(get_data(15))
```

```python
def three_buckets_pcts(df):
    cols = ["Geography",
            col("est","20K","broadband"),
            col("est","20K","NA"),
            "COMBINED $20,000 to $49,999 with broadband",
            "COMBINED $20,000 to $49,999 pop",
            col("est","75K","broadband"),
            col("est","75K","NA"),
            col("est","75+","broadband"),
            col("est","75+","NA"),
           ]
    
    return df[cols]

#three_buckets_pcts(three_buckets(get_data(2015)))
```

```python
for c in three_buckets_pcts(three_buckets(get_data(15))).columns:
    print c
```

```python
def pcts(df):
    ret = three_buckets_pcts(three_buckets(df))
    ret["pct_20K"] = ret["Estimate; $10,000 to $19,999: - With a broadband Internet subscription"]\
    / ret["Estimate; $10,000 to $19,999:"]
    ret["pct_50K"] = ret["COMBINED $20,000 to $49,999 with broadband"]\
    /ret["COMBINED $20,000 to $49,999 pop"]
    ret["pct_75K"] = ret["Estimate; $50,000 to $74,999: - With a broadband Internet subscription"]\
    /ret["Estimate; $50,000 to $74,999:"]
    ret["pct_75+"] = ret["Estimate; $75,000 or more: - With a broadband Internet subscription"]\
    /ret["Estimate; $75,000 or more:"]
    
    return ret[["Geography","pct_20K","pct_50K","pct_75K","pct_75+"]]

pcts(get_data(15))#.describe()
```
